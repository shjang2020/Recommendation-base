```python
import pandas as pd
import numpy as np
```

```python
rating_data=pd.read_csv('Ratings.csv')
```

```python
rating_data
```

```python
book_data=pd.read_csv('Books.csv')
```

```python
book_data
```

```python
rating_data['Count']=1
```

```python
df=rating_data.groupby(['ISBN'])['Count'].sum().reset_index()
```

```python
df
```

```python
df_up=df[df['Count']>=20]['ISBN'].reset_index()
```

```python
df_up
```

```python
rating=pd.merge(df_up,rating_data,left_on='ISBN',right_on='ISBN',how='left')
```

```python
rating
```

```python
rating=rating.drop(['index','Count'],axis=1)
```

```python
book=pd.merge(df_up,book_data,left_on='ISBN',right_on='ISBN',how='left')
```

```python
book.isnull().sum()
```

```python
book=book[['ISBN','Book-Title','Book-Author']]
```

```python
book=book.dropna(axis=0)
```

```python
book
```

```python
final_data=pd.merge(rating,book,on='ISBN')
final_data.head()
```

```python
final_data.shape
```

```python
final_df=final_data.pivot_table('Book-Rating',index='User-ID',columns='Book-Title').fillna(0)
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
user_rating=final_df.values.T
user_rating.shape
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
SVD=TruncatedSVD(n_components=12)
```

```python
matrix=SVD.fit_transform(user_rating)
matrix.shape
```

```python
matrix[0]
```

```python
corr=np.corrcoef(matrix)
corr.shape
```

```python
book_title=final_df.columns
book_title_list=list(book_title)
What_book=book_title_list.index('stardust')
```

```python
recommend=corr[What_book]
list(book_title[(recommend>=0.9)])[:10]
```
